```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.axhline(0, color='black', linewidth=3)
ax.axvline(0, color='black', linewidth=3)

x = [1, 2, 3, 4, 5]
y = [2, 4, 1, 6, 3]

ax.plot(x, y, marker='o', color='blue')
plt.grid(True)
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)
ax.set_xlabel("x axes")
ax.set_ylabel("y axes")
plt.show()
```
